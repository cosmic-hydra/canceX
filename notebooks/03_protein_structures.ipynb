# Protein Structure Analysis

This notebook demonstrates protein structure prediction and analysis for cancer mutations.

```python
import sys
sys.path.append('..')

from protein import ProteinStructurePredictor, StructuralAnalyzer, ProteinProteinInteractionAnalyzer
import pandas as pd
```

## Predict Protein Structure

```python
predictor = ProteinStructurePredictor(output_dir="../data/processed/structures")

# Get protein sequences
wt_sequence = predictor.get_protein_sequence('EGFR')
mut_sequence = predictor.get_protein_sequence('EGFR', mutation='L858R')

print(f"Wild-type sequence length: {len(wt_sequence)}")
print(f"Mutant sequence length: {len(mut_sequence)}")
```

```python
# Predict structures
wt_structure = predictor.predict_structure(wt_sequence, 'EGFR_WT', method='alphafold')
mut_structure = predictor.predict_structure(mut_sequence, 'EGFR_L858R', method='alphafold')

print(f"Wild-type structure: {wt_structure}")
print(f"Mutant structure: {mut_structure}")
```

## Compare Structures

```python
comparison = predictor.compare_structures(wt_structure, mut_structure)

print("Structural Comparison:")
print(f"  RMSD: {comparison['rmsd']:.2f} Å")
print(f"  TM-score: {comparison['tm_score']:.2f}")
print(f"  Significant changes: {comparison['significant_changes']}")
print(f"  Affected regions: {', '.join(comparison['affected_regions'])}")
```

## Analyze Binding Pockets

```python
analyzer = StructuralAnalyzer()

pockets = analyzer.identify_binding_pockets(mut_structure)

print(f"Found {len(pockets)} binding pockets:\n")
for pocket in pockets:
    print(f"Pocket {pocket['pocket_id']}:")
    print(f"  Volume: {pocket['volume']:.1f} Ų")
    print(f"  Druggability: {pocket['druggability_score']:.2f}")
    print(f"  Key residues: {', '.join(pocket['residues'])}\n")
```

## Analyze Mutation Impact

```python
impact = analyzer.analyze_mutation_impact(mutation_site=858, structure_file=mut_structure)

print("Mutation Impact Analysis:")
print(f"  Location: {impact['location']}")
print(f"  Stability change: {impact['stability_change']:.2f} kcal/mol")
print(f"  Predicted effect: {impact['predicted_effect']}")
print(f"  Affected regions: {', '.join(impact['functional_regions_affected']) if impact['functional_regions_affected'] else 'None'}")
```

## Predict Epitopes

```python
epitopes = analyzer.find_epitopes(mut_structure)

print(f"Predicted {len(epitopes)} epitopes:\n")
display(epitopes)
```

## Protein-Protein Interaction Analysis

```python
ppi_analyzer = ProteinProteinInteractionAnalyzer()

affected_interactions = ppi_analyzer.predict_interaction_changes(
    gene='EGFR',
    mutation='L858R',
    structure=mut_structure
)

print(f"Found {len(affected_interactions)} affected interactions:\n")
for interaction in affected_interactions:
    print(f"{interaction['partner']}:")
    print(f"  Wild-type affinity: {interaction['wild_type_affinity']:.2e} M")
    print(f"  Mutant affinity: {interaction['mutant_affinity']:.2e} M")
    print(f"  Fold change: {interaction['fold_change']:.1f}x")
    print(f"  Effect: {interaction['effect']}\n")
```

## Summary

```python
print("Protein Structure Analysis Summary:")
print(f"  Gene: EGFR")
print(f"  Mutation: L858R")
print(f"  Structural RMSD: {comparison['rmsd']:.2f} Å")
print(f"  Mutation effect: {impact['predicted_effect']}")
print(f"  Binding pockets: {len(pockets)}")
print(f"  Predicted epitopes: {len(epitopes)}")
print(f"  Affected interactions: {len(affected_interactions)}")
```
